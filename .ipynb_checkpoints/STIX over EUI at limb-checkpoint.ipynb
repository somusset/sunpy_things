This is an example of overlaying STIX contours on an EUI image in the case of part of the STIX source being off-disk (over the limb)

```python
%matplotlib notebook

import matplotlib.pyplot as plt
import astropy.units as u
from astropy.coordinates import SkyCoord
from astropy.io import fits
from astropy.time import Time
import astroquery

import sunpy.map
from sunpy.coordinates import frames, get_body_heliographic_stonyhurst, get_horizons_coord, transform_with_sun_center, Helioprojective
from sunpy.map import get_observer_meta

%matplotlib inline

plt.rcParams["figure.figsize"] = (5, 5) # Not to make image too big
```

```python
u.add_enabled_units([u.def_unit("arcsecs", 1 * u.arcsec)])
```

Load the STIX image

```python
stixdir= "data\\STIX\\"
file = stixdir+"stix_ql_image_sci_5535_uid_2111090016_16-28keV_20211109T163731_1776_clean_map.fits"
stixmap = sunpy.map.Map(file)
```

Add Solar Orbiter informaiton to the meta data of STIX

```python
solo_pos = get_horizons_coord("SOLO", stixmap.date)
stixmap.meta.update(get_observer_meta(solo_pos, rsun=solo_pos.rsun))
```

Plot the STIX image

```python
fig = plt.figure()
stixmap.plot(cmap='RdBu_r')
plt.show()
```

Load the EUI image

```python
euifile="data\\EUI\\solo_L2_eui-fsi174-image_20211109T163645293_V02.fits"
eui_map = sunpy.map.Map(euifile)
```

Plot the EUI image

```python
fig = plt.figure()
eui_map.plot()
plt.show()
```

Create a EUI sub-map, using the STIX image to know which x and y ranges to choose

```python
bottom_left = SkyCoord(900 * u.arcsec, 200 * u.arcsec, frame=eui_map.coordinate_frame)
top_right = SkyCoord(1300 * u.arcsec, 500 * u.arcsec, frame=eui_map.coordinate_frame)

sub_eui = eui_map.submap(bottom_left, top_right=top_right)
```

Plot the EUI submap to check

```python
fig = plt.figure()
sub_eui.plot()
plt.show()
```

#### This is the part that is needed for the off-disk plotting!!

We need too reproject the STIX map on the EUI submap (or whatever maps we want to overlay STIX on), using "assume_spherical_screen"

```python
with Helioprojective.assume_spherical_screen(eui_map.observer_coordinate):
    stxreproj = stixmap.reproject_to(sub_eui.wcs)
```

Choose levels for contours

```python
levels = [40.,60.,80.,90.]*u.percent
```

Plot the EUI submap and the STIX contours on top

```python
fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(projection=sub_eui)
sub_eui.plot(axes=ax, clip_interval=(1, 99.99)*u.percent)
sub_eui.draw_grid(axes=ax)

ax.set_title("STIX over EUI", y=1.1)

cset = stxreproj.draw_contours(levels, axes=ax, alpha=0.9, colors='blue')
#ax.axis(bounds)

plt.show()
```
