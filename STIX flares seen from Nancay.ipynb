## Select events seen from Nancay

```python
from datetime import datetime, date, time
```

First, select events seen from Earth

```python
Flares_seen_from_Earth = stix_flarelist[stix_flarelist["hpc_x_earth"].notnull()]
```

```python
Flares_seen_from_Earth
```

Now select flares that happenend during Nancay observing time

Define observing times

```python
nrh_start = time(9)
nrh_end = time(16)
```

Extract STIX times

```python
flare_start_time = pd.to_datetime(Flares_seen_from_Earth["start_UTC"]).dt.time.values
flare_end_time = pd.to_datetime(Flares_seen_from_Earth["end_UTC"]).dt.time.values
```

Down select the flares

```python
Flares_seen_from_Nancay = Flares_seen_from_Earth[(flare_start_time > nrh_start)&(flare_end_time < nrh_end)]
```

```python
Flares_seen_from_Nancay
```

## Visualisation of Nancay-STIX flare list

Plot flare positions seen from Earth

```python
import matplotlib.pyplot as plt
from matplotlib import dates
```

Create tick marks with the time of the flares

```python
ticks = dates.date2num(pd.to_datetime(Flares_seen_from_Nancay["peak_UTC"]))
```

Create sizes that matches the size of the flares (GOES flux)

```python
sizes = Flares_seen_from_Nancay["GOES_flux_time_of_flare"].values*4e6
```

```python
fig, ax = plt.subplots(figsize=(8, 8), dpi=200)
im = plt.scatter(Flares_seen_from_Nancay["hpc_x_earth"], Flares_seen_from_Nancay["hpc_y_earth"], sizes, c=ticks, alpha=0.7)
ax.set_aspect('equal', adjustable='box')
ax.set_xlim(-1200, 1200)
ax.set_ylim(-1200, 1200)
circle1 = plt.Circle((0, 0), 960, color='grey', fill=False)
ax.add_artist(circle1)
cbar = fig.colorbar(im, ticks=[ticks[0],ticks[int(len(ticks)/2)], ticks[-1]], pad=0.1, shrink=0.8)
cbar.ax.set_yticklabels([dates.num2date(ticks[0]).strftime("%Y-%m-%d"), 
                        dates.num2date(ticks[int(len(ticks)/2)]).strftime("%Y-%m-%d"),
                        dates.num2date(ticks[-1]).strftime("%Y-%m-%d")])  # vertically oriented colorbar
```
